# visualise higher band images

```python
import os
import rasterio
from rasterio.plot import reshape_as_image
import matplotlib.pyplot as plt

MS_loc = "/home/titanx/Desktop/dataset/Mumbai/MS"
img_name = "SN5_roads_train_AOI_8_Mumbai_MS_chip8.tif"
MS_img = os.path.join(MS_loc, img_name)

with rasterio.open(MS_img) as src:
    # NIR1 (7), Red (5), Green (3)
    image = src.read([8, 1, 3])
    
    # Reshape to (H, W, 3)
    image_reshaped = reshape_as_image(image)
    
    # Normalize the pixel values for plotting
    image_reshaped = (image_reshaped - image_reshaped.min()) / (image_reshaped.max() - image_reshaped.min())
    
    plt.figure(figsize=(6, 6))
    plt.imshow(image_reshaped)
    plt.title("Standard False Color (NIR, Red, Green)")
    plt.axis('off')
    plt.show()
```

# visualise RGB image

```python
import os
import matplotlib.pyplot as plt 
import rasterio
from rasterio.plot import reshape_as_image
import numpy
```

```python
img_dir = "/home/titanx/Desktop/dataset/Mumbai/PS-RGB"
img_path = os.path.join(img_dir, "SN5_roads_train_AOI_8_Mumbai_PS-RGB_chip8.tif")

with rasterio.open(img_path) as src:
    image = src.read([1, 2, 3]) 
    
    image_reshaped = reshape_as_image(image)
    
    if image_reshaped.max() > 1.0:
        image_reshaped = (image_reshaped - image_reshaped.min()) / (image_reshaped.max() - image_reshaped.min())
    
    plt.imshow(image_reshaped)
    plt.show()
```

```python
import cv2

def check_bit_depth_opencv(image_path):
    # IMREAD_UNCHANGED is crucial, otherwise OpenCV converts 16-bit to 8-bit automatically
    img = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)
    
    if img is None:
        print("Error: Could not load image.")
        return

    # Check the data type of the underlying NumPy array
    if img.dtype == 'uint8':
        print(f"{image_path}: 8-bit image")
    elif img.dtype == 'uint16':
        print(f"{image_path}: 16-bit image")
    else:
        print(f"{image_path}: Other data type ({img.dtype})")

# Example usage
check_bit_depth_opencv('/home/titanx/Desktop/dataset/Mumbai/PS-RGB/SN5_roads_train_AOI_8_Mumbai_PS-RGB_chip0.tif')
```
